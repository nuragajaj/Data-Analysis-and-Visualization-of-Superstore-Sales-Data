```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
dataset = pd.read_excel("Superstore_usa.xlsx")
dataset.head(24)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
# it check or remove null value
# Fill NaN values in 'Product Base Margin' column with the mean
mean_margin = dataset['Product Base Margin'].mean()
dataset['Product Base Margin'].fillna(mean_margin, inplace=True)
```

```python
dataset['Order Priority'].value_counts()
```

customer segment

```python
sns.countplot(x='Customer Segment',data = dataset)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x = 'Order Priority',data = dataset)
plt.title('Count of Order Priority')
plt.show()
```
